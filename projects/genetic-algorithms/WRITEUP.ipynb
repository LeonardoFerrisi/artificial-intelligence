# The HillClimber Algorithm

Changes seeds by mutating each bitstring with a Mutation Rate of 0.005, for every bitstring in a population of 128 per run
The algorithm repeats this process until it reaches a maximum fitness value of 256 (**matches with all components of royal road (including intermediate)**)

#### Plotting the data

The graph below depicts statistics showing the aggregation of 30 runs of the hillclimber algroithm with metrics all derived from the maximums in every run.

The Maximim value is the maximum fitness of all 30 runs for each generation. (Example: Max Fitness at Generation 5 is simply the maximum of all the MAX values at gen 5 for all 30 runs)

The Mean value is the mean of the maximum fitness of all 30 runs for each generation. (Example: Mean Max Fitness at Generation 5 is simply the mean of all the MAX values at gen 5 for all 30 runs)

The Min value is the Minimum of all max fitness values for all 30 runs for each gen. (Example: Minimum Max Fitness at Generation 5 is simply the min of all the MAX values at gen 5 for all 30 runs)
