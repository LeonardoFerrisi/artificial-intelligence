The following graphs show the two conditions of QLearning with and without Decay implemented, conducted 30 times with 20,000 episodes per run. 

```python
plotQLearn('q1', title="Q Learning on CoffeeGame", yLabel='Average Reward (per 1000 Episodes)', xLabel='Num Episodes')
```

### Analysis:

The Algorithm seems to consistently increase in average reward and then sliightly decrease approaching a consistent range between 1.0 and 0.8 (where average reward was total reward/number of steps) and plotted with the average of every 100 episodes for 20 episodes.

This QLearning program eventually, across 20,000 episodes reaches a point where it consistently can reach the goal state in the minimum amount of steps. In the case of the coffee came this is 4 steps. (Right, Right, Down, Down)

s[0] Start -->  Right --> s[1] Right --> s[2] Down --> s[5] Down --> s[8] Goal

There are other 4 step paths to reach goal such as (Right, Down, Right, Down) which accounts for some of the variability in results.

Without a decay rate, the exploration Rate (epsilon) stays at a solid 0.5 the entire time, meaning that the action policy generated will favor the best move in any given state 50% of time. 



```python
plotQLearnWDecay('q2', title="Q Learning with Decay on Taxi V3", yLabel='Average Reward (per 1000 Episodes)', xLabel='Num Episodes')
```

### Analysis:

The Algorithm seems to initally dip negatively, likely because the exploration rate is still high enough that the our Reinforcement learner has some time to hit highly negative squares before the exploration rate decays enough such that the learner then consistently increase in average reward approaching a consistent cap of 1 (where average reward was total reward / steps taken) and plotted with the average of every 100 episodes for 20 episodes.

This QLearning program eventually, across 20,000 episodes reaches a point where it consistently can reach the goal state in the minimum amount of steps. In the case of the coffee came this is 4 steps. (Right, Right, Down, Down)

s[0] Start -->  Right --> s[1] Right --> s[2] Down --> s[5] Down --> s[8] Goal

There are other 4 step paths to reach goal such as (Right, Down, Right, Down) which accounts for some of the variability in results.

As Episilon approaches 0, the total reward / steps taken per episode tends toward 1.00.
